```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_movies.csv') 
```

```python
movies.head(1)
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
movies = movies[['id_x','title','movie_name_x','overview_x','genres_x','keywords_x','vote_average_y','vote_count_y']]
```

```python
movies.head(1)
```

```python
movies.describe()
```

```python
movies.info()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres_x
```

```python
movies['genres_x'] = movies['genres_x'].apply(convert)
movies.head()
```

```python
movies['keywords_x'] = movies['keywords_x'].apply(convert)
movies.head()
```

```python
def transform(string):
    li = list(string.split(" "))
    return li
```

```python
transform('movie_name_x')
movies['movie_name_x'] = movies['movie_name_x'].apply(transform)
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['genres_x'] = movies['genres_x'].apply(collapse)
movies['keywords_x'] = movies['keywords_x'].apply(collapse)
movies['movie_name_x'] = movies['movie_name_x'].apply(collapse)
```

```python
movies['overview_x'] = movies['overview_x'].apply(lambda x:x.split())
movies.head()
```

```python
movies['tags'] = movies['movie_name_x'] + movies['overview_x'] + movies['genres_x'] + movies['keywords_x']
```

```python
new = movies.drop(columns=['overview_x','genres_x','keywords_x','movie_name_x'])
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vector).shape
```

```python
similarity = cosine_similarity(vector)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[0:10]
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])[0:10]
    for i in distances[0:10]:
        print(new.iloc[i[0]].title)


    
```

```python
recommend('The Avengers')
```
